```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('hiring.csv')
```

```python
data
```

```python
data['salary($)'] = data['salary($)'].fillna(data['salary($)'].mean())
```

```python
data
```

```python
data = data.fillna({'experience' : 'three'})
```

```python
data
```

```python
data.isnull().sum()
```

```python
data
```

```python
data.experience.value_counts()
```

```python
data['experience'] = data['experience'].map({'three':3, 'ten':10, 'seven':7, 'five':5, 'two':2, 'eleven':11})
```

```python
data
```

```python
x = data[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
```

```python
x
```

```python
y = data[['salary($)']]
```

```python
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
p = reg.predict(x)
```

```python
p
```

```python
data['salary($)'] = p
```

```python
data
```

```python
data.to_csv('salary_prediction.csv',index=False)
```
